**Read data as X_train, y_train & X_test**

Check for missing values and take features that could be useful in determining house pricing.

Apply one hot encoding on categorical values.

```python
X_df = pd.read_csv('/kaggle/input/house-prices-advanced-regression-techniques/train.csv')
X_df.head()
```

```python
y_train = X_df['SalePrice']
y_train.head()
```

```python
X_df.isna().sum()
```

**Features chosen:**

MSSubClass: Type of housing

MSZoning: The zone the house is located in

LotArea: Lot size

Utlities: Type of utilities house has

OverallQual: Overall quality of the house

OverallCond: Overall condition of the house

1stFlrSF: Area of the first floor

2ndFlrSF: Area of the 2nd floor

Functional: Functionality/Quality of the house for living in

YrSold: Year the house is sold


```python
features = ['MSSubClass', 'MSZoning', 'LotArea', 'Utilities', 'OverallQual', 'OverallCond', '1stFlrSF', '2ndFlrSF', 'Functional', 'YrSold']
X_train_ohe = X_df[features]
X_train_ohe.head()
```

```python
categorical_cols = ['MSZoning', 'Utilities', 'Functional']
X_train = pd.get_dummies(X_train_ohe, columns = categorical_cols)
X_train.head()
```

```python
X_train = X_train.to_numpy()
print(X_train[:5])
```

Training data has 1 or more houses with the categorical value 'NoSeWa' under Utilities, while the test data does not.
Added Utilities_NoSeWa column with all false values to fit with the training data size.

```python
test_df = pd.read_csv('/kaggle/input/house-prices-advanced-regression-techniques/test.csv')
X_test = test_df[features]

X_test = pd.get_dummies(X_test, columns = categorical_cols)
X_test.insert(13, "Utilities_NoSeWa", False)
print(X_test.head())
X_test = X_test.to_numpy()
```

**Sklearn linear regression to fit training data & predict housing price results**

```python
reg = LinearRegression()
reg.fit(X_train, y_train)
reg.score(X_train, y_train)
```

```python
predict = reg.predict(X_test)
print(predict[:10])
```

```python
output = pd.DataFrame({'Id':test_df.Id, 'SalePrice':predict})
output.to_csv('submission.csv', index = False)
output.head()
```
